```python
import numpy as np
```

```python
x = np.arange(0,10)
```

```python
y = np.arange(11,21)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y,c= "r")
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title("graph in 2D")
```

```python
y = x*x
```

```python
plt.plot(x,y,"bo")
```

```python
plt.plot(x,y,"b*",linestyle= "dashed")
```

```python
plt.plot(x,y,"b--",linestyle= "dashed")
```

```python
plt.plot(x,y,"b*",linestyle= "dashed",linewidth=0.5)
```

```python
plt.plot(x,y,"b*",linestyle= "dashed",linewidth=0.5,markersize=5)
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

CREATING SUBPLOT


```python
plt.subplot(2,2,1)
plt.plot(x,y,"r--")
plt.subplot(2,2,2)
plt.plot(y,x,"g--")
plt.subplot(2,2,3)
plt.plot(x,y,"b--")
plt.subplot(2,2,4)
plt.plot(x,y,"b--")
#Assignment x title and y title
```
